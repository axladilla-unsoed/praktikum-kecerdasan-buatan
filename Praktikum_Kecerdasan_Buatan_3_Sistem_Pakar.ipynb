## **Dasar Teori**


### **1. Sistem Pakar**
* Sistem pakar adalah sebuah sistem yang mengadopsi pengetahuan manusia ke dalam
komputer, sehingga komputer dapat menangani masalah tertentu dengan meniru metode kerja ahli.
* Sistem pakar memungkinkan transfer keahlian dari seorang ahli (atau sumber lainnya) ke komputer.
* Informasi tersebut disimpan dalam komputer, dan pengguna dapat berkonsultasi
untuk mendapatkan saran. Komputer kemudian dapat membuat kesimpulan, layaknya
seorang ahli, serta memberikan penjelasan dan alasan ketika diperlukan
* Sistem pakar meniru cara berpikir ahli dengan:
  * Knowledge Base (basis pengetahuan)
  * Inference Engine (mesin inferensi)

````python
# Knowledge base (aturan sederhana)
rules = {
    "demam": "Kemungkinan Flu",
    "batuk": "Kemungkinan ISPA",
    "pusing": "Kemungkinan Migrain"
}

# Input user
gejala = "demam"

# Inferensi sederhana
if gejala in rules:
    print("Diagnosis:", rules[gejala])
else:
    print("Tidak diketahui")
````

````python
"""
Buat file gejala.pl yang berisi

%Knowledge base (aturan)
aturan(demam, 'Kemungkinan Flu').
aturan(batuk, 'Kemungkinan ISPA').
aturan(pusing, 'Kemungkinan Migrain').

%Rule inferensi
diagnosa(X, Y) :- aturan(X, Y).
"""

from pyswip import Prolog

prolog = Prolog()
prolog.consult("gejala.pl")

# Input user
gejala = "demam"

# Query ke Prolog
query = list(prolog.query(f"diagnosa({gejala}, Y)", maxresult=1))
print(query)
````

### **2. Logika Predikat**
* Logika predikat (kalkulus predikat) merupakan bagian dari komputasi logika  yang juga mencakup aljabar Boole (logika proporsional), di mana fakta dan aturan dinyatakan melalui predikat seperti berikut:
```
lelaki(Joko)                                    // fakta
menikah(Joko, Tuti)                             // fakta
∀x ∀y [menikah(x,y) ∧ lelaki(x)] → ~lelaki(y)   // aturan; ∀ berarti "untuk setiap"
∀y ∃x [orang(y) → ibu(x,y)]                     // aturan; ∃ berarti "ada"
```

* Kalimat pertama menunjukkan adanya fakta bahwa Joko adalah seorang lelaki
* Kalimat kedua menyatakan fakta bahwa Joko menikah dengan Tuti.
* Kalimat ketiga dan keempat menunjukkan suatu aturan atau kaidah yang berlaku.
  * Kalimat ketiga menyatakan aturan
  ```
  bahwa untuk setiap pasang orang x dan y,
  jika x menikah dengan y dan x adalah lelaki,
  dapat disimpulkan bahwa y adalah bukan seorang lelaki.
  ```
  * Kalimat keempat menyatakan aturan
  ```
  bahwa untuk setiap y,
  jika y adalah orang,
  y mempunyai seorang ibu x (x adalah ibu dari y).
  ```

* Simbol predikat yang digunakan dalam kalimat tersebut adalah
  * lelaki, menikah, orang, dan ibu yang dapat disebut sebagai simbol relasi
  * Joko dan Tuti yang dapat disebut sebagai simbol konstanta.

Contoh Lain :

```
bapak(alya, bima)
bapak(bima, david)
kakek(A,Z) :- bapak(A,X), bapak(X,Z)
```

Dimana berarti

bapak(alya, bima) & bapak(bima, david) adalah fakta

kakek(A,Z) adalah aturan dimana
```
A adalah kakek Z jika A adalah bapak dari X dan X bapak dari Z
```

````python
"""
Buat file kakek.pl yang berisi

%Knowledge base (aturan)
bapak(alya, bima).
bapak(bima, david).

%Rule inferensi
kakek(X,Y) :- bapak(X,Z), bapak(Z,Y).
"""

from pyswip import Prolog

prolog = Prolog()
prolog.consult("kakek.pl")

# Query ke Prolog
query = list(prolog.query("kakek(alya, Y)"))
print(query)
````

### **3. Bahasa Deklaratif & Prolog**

* Bahasa pemrograman biasanya bersifat prosedural atau imperatif, di mana program disusun melalui langkah-langkah dan prosedur yang terstruktur.
* Namun, ada juga bahasa pemrograman yang bersifat deskriptif atau deklaratif, seperti Prolog.
* Prolog adalah singkatan dari Programming in Logic yang berarti pemrograman logika. Prolog menggunakan bahasa deklaratif, di mana pemrogram memberi fakta dan aturan untuk selanjutnya diselesaikan oleh Prolog secara deduktif sehingga menghasilkan suatu kesimpulan.
* Selain itu, berbeda dengan pemrograman fungsional, pemrograman logika
menggunakan relasi, bukan fungsi, sehingga sangat sesuai untuk mengimplementasikan sebuah sistem pakar.

Sebagai contoh, dalam Prolog, bahasa deklaratif digunakan untuk menyatakan fakta dan aturan sebagai berikut:

* Jika ingin menyatakan bahwa Prawiro adalah bapak dari Joko, dalam Prolog
dituliskan sebagai berikut
  * bapak(prawiro, joko).
* Jika ingin menerangkan suatu kaidah bahwa A adalah kakek dari Z, dalam Prolog
dituliskan sebagai berikut:
  * Aturan dalam Bahasa Indonesia:
    * A adalah kakek dari Z jika A adalah bapak dari X dan X adalah bapak dari Z.
    * A adalah kakek dari Z jika A adalah bapak dari X dan X adalah ibu dari Z.
  * Aturan dalam Bahasa Prolog:
    * kakek(A,Z) :- bapak(A,X), bapak(X,Z).
    * kakek(A,Z) :- bapak(A,X), ibu(X,Z).

````python
"""
Buat file keluarga.pl yang berisi

bapak(prawiro, joko).
bapak(joko, budi).
ibu(joko, siti).

kakek(A,Z) :- bapak(A,X), bapak(X,Z).
kakek(A,Z) :- bapak(A,X), ibu(X,Z).
"""

from pyswip import Prolog

prolog = Prolog()
prolog.consult("keluarga.pl")

print(list(prolog.query("kakek(prawiro, Z)")))
````

### **4. Dasar Pemrograman Prolog**

Pada bagian ini akan diuraikan dasar-dasar pemrograman Prolog, aturan umum penulisan program, bagaimana melakukan dialog dengan Prolog, dan beberapa pengertian dasar yang berkaitan dengan program Prolog.

#### **a. Klausa**

* Setiap program Prolog tersusun dari klausa-klausa.
* Klausa dapat berupa fakta atau aturan.
* Klausa terdiri dari head (kepala) dan body (tubuh) yang dipisahkan dengan tanda `:-` dan diakhiri dengan tanda titik `.`
```
diagnosa(X, Y) :- aturan(X, Y).
```

#### **b. Fakta**

* Fakta adalah suatu kenyataan atau kebenaran yang diketahui atau hubungan
(relasi) antara dua atau lebih objek.
* Fakta dapat pula menunjukkan sifat suatu objek.
* Klausa fakta hanya terdiri dari head saja

Contoh sederhana adalah:

* bapak(prawiro, joko).
* merah(darah).
* asin(garam).

**Fakta (Hanya Terdiri dari Head)**

```
lelaki(joko).
perempuan(siti).

bapak(prawiro, joko).
ibu(joko, siti).
```

#### **c. Aturan**

* Aturan merupakan logika yang dirumuskan dalam bentuk relasi sebab-akibat dan
hubungan implikasi.
* Misalnya, dapat dibuat aturan bahwa jika A adalah bapak dari X dan X adalah bapak atau ibu dari Z, dapat dipastikan bahwa A adalah kakek dari Z.

**Aturan (head :- body)**

```
anak(X,Y) :- bapak(Y,X).
```

Penjelasan:

Head → anak(X,Y)

Body → bapak(Y,X)

Artinya:

X adalah anak dari Y jika Y adalah bapak dari X

#### **d. Penggabungan Klausa**

* Tubuh klausa dapat terdiri dari beberapa sub-klausa yang dihubungkan satu
sama lain menggunakan tanda:
  * koma (,) yang menunjukkan hubungan AND
  * tanda titik koma (;) yang menunjukkan hubungan OR.
* Penggabungan tubuh klausa
  * Penggabungan tubuh klausa yang dirangkai dengan AND disebut sebagai konjungsi
  * Penggabungan tubuh klausa yang dirangkai dengan OR disebut sebagai disjungsi.

**Konjungsi (AND / `,` )**

```
kakek(A,Z) :- bapak(A,X), bapak(X,Z).
```

* Penjelasan:

Body terdiri dari 2 sub-klausa

`bapak(A,X)`

`bapak(X,Z)`

* Artinya:
```
A adalah kakek dari Z jika
A bapak dari X DAN X bapak dari Z
```

**Disjungsi (OR / `;` )**

```
orang_tua(X,Y) :- bapak(X,Y) ; ibu(X,Y).
```

* Artinya:
```
X adalah orang tua dari Y jika
X bapak dari Y ATAU X ibu dari Y
```

**Kombinasi (AND dan OR)**

```
kakek(A,Z) :- bapak(A,X), (bapak(X,Z) ; ibu(X,Z)).
```
atau jika dijabarkan
```
kakek(A,Z) :- bapak(A,X), bapak(X,Z).
kakek(A,Z) :- bapak(A,X), ibu(X,Z).
```

* Penjelasan:

`bapak(A,X) merupakan syarat pertama`

`(bapak(X,Z) ; ibu(X,Z)) merupakan perintah OR`

Artinya:
```
A adalah kakek dari Z jika
A bapak dari X DAN
(X bapak dari Z ATAU X ibu dari Z)
```

#### **e. Predikat**

* Kumpulan klausa dengan struktur head yang sama disebut sebagai predikat.
* Suatu predikat memiliki nama dan nol atau lebih argumen.
* Nama predikat berupa atom, sedangkan argumen predikat dapat berupa atom atau variabel.

```
% Fakta (knowledge base)
bapak(prawiro, joko).
bapak(joko, budi).
ibu(joko, siti).

% Predikat kakek/2 (terdiri dari beberapa klausa)
% Klausa 1
kakek(A,Z) :- bapak(A,X), bapak(X,Z).

% Klausa 2
kakek(A,Z) :- bapak(A,X), ibu(X,Z).
```

Pada aturan kakek, terdapat dua klausa yang menyatakan aturan bagaimana
syarat A adalah kakek dari Z. `Kakek(A,Z)` tersebut adalah predikat yang menerima dua parameter, atau dapat ditulis sebagai `kakek/2.`

#### **f. Atom dan Variabel**


**Atom**

* Argumen suatu predikat dapat berupa konstanta (atom) atau variabel. Atom
disebut juga sebagai objek nyata, sedangkan variabel disebut sebagai objek umum.
* Suatu atom atau variabel dalam Prolog disebut sebagai term, sehingga argumen suatu predikat selalu berupa term.
```
bapak(prawiro, joko).
```
'prawiro' dan 'joko' adalah atom (objek nyata)

**Variabel**
* Dalam Prolog, setiap term yang ditulis dengan awalan huruf kapital adalah
variabel bernama, sedangkan awalan dengan huruf kecil adalah atom atau konstanta.
```
anak(X, Y) :- bapak(Y, X).
```
X, Y → variabel (huruf kapital)

bisa merepresentasikan nilai apa saja

**Aturan penamaan**
```
% BENAR
bapak(prawiro, joko).   % atom (huruf kecil)
bapak(X, Y).            % variabel (huruf kapital)

% SALAH (akan dianggap variabel)
bapak(Prawiro, joko).   % 'Prawiro' dianggap variabel!
```


**Variabel tak bernama (Anonymous Variable)**
* Variabel tak bernama digunakan untuk mengabaikan nilai suatu variabel, yang berarti bisa bernilai apa saja.
```
% siapa saja yang punya anak (tidak peduli siapa anaknya)
punya_anak(X) :- bapak(X, _).
```
Artinya

* tidak peduli nilainya
* tidak disimpan
* tidak di-track oleh Prolog

#### **g. Query**
* Query atau Pertanyaan digunakan untuk memperoleh jawaban dari suatu masalah
secara deduktif berdasarkan fakta-fakta dan aturan-aturan yang telah ditentukan.
* Dalam Prolog, query dinyatakan dengan goal.
* Goal terdiri dari dua jenis, yaitu
    * Goal internal yang ditulis langsung di dalam program
    * Goal eksternal yang ditulis di luar program atau di console.

Berikut adalah beberapa contoh goal:



**Goal Internal**
```
orangtua(P,Q) :- bapak(P,Q); ibu(P,Q).
```
  * Pada klausa tersebut, terdapat dua goal internal, yaitu
`bapak(P,Q)` dan `ibu(P,Q)`
* di mana goal pertama menanyakan program apakah `P adalah bapak dari Q` dan goal kedua menanyakan program apakah `P adalah ibu dari Q`.

**Goal Eksternal**


  * **Menampilkan Output***
  ```
  write("Hello World"), nl.
  ```
  * Pada klausa tersebut, terdapat dua goal eksternal, yaitu write dan nl.
  * Memasukkan query tersebut di console akan menampilkan tulisan “Hello World” dan membuat satu baris baru.
  

  * **Query fakta (boolean)**
  ```
  bapak(prawiro, joko).
  ```
  Menanyakan:
  
  apakah benar prawiro adalah bapak dari joko?

  * **Query dengan variabel (mencari jawaban)**
  ```
  bapak(X, joko).
  ```
  Menanyakan:
  
  siapa bapak dari joko?

````python
"""
Buat file query_fakta.pl yang berisi

bapak(prawiro, joko).
bapak(prawiro, setiadi).
"""

from pyswip import Prolog

prolog = Prolog()
prolog.consult("query_fakta.pl")

print(list(prolog.query("bapak(X, joko)")))
````

  ```
  bapak(prawiro, X).
  ```
  Menanyakan:
  
  siapa anak dari prawiro?

````python
"""
Buat file query_fakta.pl yang berisi

bapak(prawiro, joko).
bapak(prawiro, setiadi).
"""

from pyswip import Prolog

prolog = Prolog()
prolog.consult("query_fakta.pl")

print(list(prolog.query("bapak(prawiro, X)")))
````

  * bapak(X, joko).

  * bapak(prawiro, X).

## **Percobaan 1 - Silsilah Keluarga**



````python
"""
Buat file silsilah_keluarga.pl yang berisi

% SILSILAH KELUARGA
% FAKTA
% -- Hubungan orang tua.
% -- Predikat parent/2 (menerima dua parameter X dan Y)
% -- berarti X adalah orang tua dari Y.
parent(alya, bima).
parent(alya, satria).
parent(bima, david).
parent(bima, emma).
parent(satria, yunita).
parent(satria, grace).

% ATURAN
% -- Hubungan saudara kandung.
% -- Predikat sibling/2 (menerima dua parameter X dan Y)
% -- berarti X adalah saudara kandung Y apabila fakta-fakta sesuai dengan
aturan.

% -- Hubungan kakek-nenek.
% -- Predikat grandparent/2 (menerima dua parameter X dan Y)
% -- berarti X adalah kakek/nenek dari Y apabila fakta-fakta sesuai
dengan aturan.

grandparent(X, Y) :- parent(X, Z),parent(Z, Y).
% -- Hubungan turun-temurun.
% -- Predikat ancestor/2 (menerima dua parameter X dan Y)
% -- berarti X adalah pendahulu Y apabila fakta-fakta sesuai dengan aturan.

ancestor(X, Y) :- parent(X, Y).
ancestor(X, Y) :- parent(X, Z), ancestor(Z, Y).
"""

from pyswip import Prolog

prolog = Prolog()
prolog.consult("silsilah_keluarga.pl")

print(list(prolog.query("parent(bima, Anak)"))) # Siapa anaknya Bima?
print(list(prolog.query("grandparent(alya, Cucu)"))) # Siapa cucunya Alya?
print(list(prolog.query("ancestor(Leluhur, emma)"))) # Siapa leluhurnya Emma?
````

````python
"""
Buat file penyakit.pl yang berisi

% GEJALA PENYAKIT MALARIA
% DATABASE
% -- Menyimpan data apakah X memiliki gejala Y.
:- dynamic gejala/2.

% ATURAN
% -- Mengandung daftar gejala untuk beberapa jenis penyakit malaria,
% -- dengan predikat penyakit/2 (menerima dua parameter)
% -- yang berarti X mengidap penyakit Y.
% -- Predikat gejala/2 berarti X memiliki gejala Y.
penyakit(X, tertiana) :-
gejala(X, nyeri_otot),
gejala(X, muntah),
gejala(X, kejang).

penyakit(X, quartana) :-
gejala(X, nyeri_otot),
gejala(X, menggigil),
gejala(X, tidak_enak_badan).

penyakit(X, tropika) :-
gejala(X, keringat_dingin),
gejala(X, sakit_kepala),
gejala(X, mimisan),
gejala(X, mual).

penyakit(X, pernisiosa) :-
gejala(X, menggigil),
gejala(X, tidak_enak_badan),
gejala(X, demam),
gejala(X, mimisan),
gejala(X, mual).



"""

from pyswip import Prolog

prolog = Prolog()

Prolog.retractall("gejala(_, _)") # untuk mereset memori

prolog.consult("penyakit.pl")
````

````python
print(list(prolog.query("penyakit(steph, Penyakit)")))
````
